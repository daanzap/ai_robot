```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimage
import numpy as np

dogs_and_fish = mpimage.imread('fish_and_dogs.png')
imgplot = plt.imshow(dogs_and_fish, cmap='gray')
def over_lay(weights, org_img=None):
    if org_img is None:
        org_img = dogs_and_fish
    for row in range(0,199):
        for column in range(0,199):
            if org_img[row][column] > 0.5:
                
                weights[row][column] = org_img[row][column] * 2
    return weights
```

```python

def make_divider(W,B):
    divider = np.zeros((200,200))
    for row in range(0,199):
        for column in range(0,199):
            divider[row][column] = np.tanh((column * W + (B+100) - row)/40.)
                                       
    return np.clip(divider, 0, 5)

divider = make_divider(0,0)
imgplot = plt.imshow(over_lay(divider), cmap='gray')
```

```python
divider2 = make_divider(0.5,-30)
imgplot = plt.imshow(over_lay(divider2), cmap='gray')
```

```python
sum_divider = (divider)+(divider2)
imgplot = plt.imshow(over_lay(sum_divider), cmap='gray')
```

```python
divider3 = make_divider(-1,40)
imgplot = plt.imshow(over_lay(divider3), cmap='gray')
```

```python
divider4 = make_divider(0,-70)[::-1]
imgplot = plt.imshow(over_lay(divider4), cmap='gray')
```

```python
sum_divider =(divider*0.5)+(divider2*5)+(divider3 * 5)+(divider4* 10)
imgplot = plt.imshow(over_lay(sum_divider), cmap='gray')
```

```python
sum_divider =np.clip((divider*0.5)+(divider2*5)+(divider3 * 5)+(divider4* 10),0.,0.4)
imgplot = plt.imshow(over_lay(sum_divider), cmap='gray')
```
